```python
import pandas as pd
income= pd.read_csv('income.csv')
income.head()
```

```python
income['Income($)'].min()
```

```python
income['Income($)'].max()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
st_scaler= StandardScaler()
```

```python
income.loc[:,'Income($)'].std()
```

```python
# z = (x - u) / s
s= income.loc[:,'Income($)'].std()
u= income.loc[:,'Income($)'].mean()

z= (90000-u)/s
z
```

```python
income.loc[:,'new_standard']= st_scaler.fit_transform(income.loc[:,['Income($)']])
```

```python
income.head()
```

**MinMax Scaler**

```python
# X_std = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
# X_scaled = X_std * (max - min) + min
minmax_scaler= MinMaxScaler(feature_range=(0,4)) #feature_range=(0,1)
```

```python
income.loc[:,'Minmax_values']= minmax_scaler.fit_transform(income.loc[:,['Income($)']])
```
