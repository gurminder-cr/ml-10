```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
df= pd.read_csv('income.csv')
df.head()
```

```python
df= df.iloc[:,1:]
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
fig= plt.scatter(x=df['Age'],y=df['Income($)'])
plt.show()
```

```python
kmeans= KMeans(n_clusters=3)
val= kmeans.fit_predict(df[['Age','Income($)']])
```

```python
val
```

```python
df.loc[:,'target']=val
```

```python
df
```

```python
center = kmeans.cluster_centers_
center
```

```python
center[:,0]
```

```python
plt.scatter(df[df.loc[:,'target']==0]['Age'], df[df.loc[:,'target']==0]['Income($)'])
plt.scatter(df[df.loc[:,'target']==1]['Age'], df[df.loc[:,'target']==1]['Income($)'])
plt.scatter(df[df.loc[:,'target']==2]['Age'], df[df.loc[:,'target']==2]['Income($)'])
plt.scatter(center[:,0], center[:,1],marker='*' )
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler= MinMaxScaler()
df.loc[:,['Age','Income($)']]= scaler.fit_transform(df.loc[:,['Age','Income($)']])
```
